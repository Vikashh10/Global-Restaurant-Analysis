```python
#Check missing values
df.isnull().sum()
```

```python
#Handling missing values
df = df.dropna(subset=["Cuisines"])
```

```python
#Confirm again
df.isnull().sum()
```

```python
#Check for duplicates
df.duplicated().sum()
```

```python
#Exploratory Data Analysis and Visualization
```

```python
#Top cities with most restaurants
top_cities = df["City"].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.values,y=top_cities.index, palette="viridis")
plt.title("Top 10 Cities with most Restaurants")
plt.xlabel('Number of Restaurants')
plt.ylabel("City")
plt.show()
```

```python
#Distribution of ratings
plt.figure(figsize=(8,5))
sns.histplot(df["Aggregate rating"], bins=20, kde=True, color="orange")
plt.title("Rating Distribution")
plt.xlabel("Aggregate rating")
plt.ylabel("Count")
plt.show()
```

```python
#Average cost for two by price range

plt.figure(figsize=(8,5))
sns.boxplot(x="Price range", y="Average Cost for two", data=df)
plt.title("Cost for Two by Price Range")
plt.show()
```

```python
#Online Delivery Availability

online_delivery = df["Has Online delivery"].value_counts()
plt.figure(figsize=(6,4))
plt.pie(online_delivery, labels=["No","Yes"], autopct="%1.1f%%", colors=["yellow","green"], startangle=90)
plt.title("Online Delivery Availability")
plt.show()
```
